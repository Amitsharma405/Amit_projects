QUES.1 Explain the importance of Functions?


Functions are important because they help us organize, reuse, and understand how things work—especially in math, programming, and science. Here’s why they matter:

1. Break big problems into smaller pieces

Functions let you take a large, complicated problem and split it into smaller, manageable parts. Each function handles one specific task, which makes problem-solving much easier.

2. Reusability

Once you define a function, you can use it again and again without rewriting the same steps. This saves time, reduces effort, and avoids mistakes.

3. Clarity and readability

Functions give meaningful names to actions (like calculateArea or findAverage). This makes code, formulas, or processes easier to read and understand—for you and for others.

4. Easier testing and debugging

If something goes wrong, you can test or fix one function at a time instead of searching through an entire program or process.

5. Abstraction

Functions hide complex details. You only need to know what a function does, not how it does it. This helps you focus on the bigger picture.

6. Consistency and accuracy

Using the same function ensures the same logic is applied every time, which reduces errors and keeps results consistent.

7. Real-world modeling

In math and science, functions describe relationships between quantities (like speed vs. time or cost vs. distance), helping us analyze and predict real-world behavior.

ques.10 Write a code that generates prime numbers up to n using a generator?

```python
def prime_numbers(n):
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num

n = 30
for prime in prime_numbers(n):
    print(prime)
```

ques.48 Implement a class BankAccount with a private balance attribute. Include methods to deposit, withdraw,
and check the balance?

```python
class BankAccount:
    def __init__(self, initial_balance=0.0):
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative")
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount

    def get_balance(self):
        return self.__balance
account = BankAccount(100)

account.deposit(50)
account.withdraw(30)

print(account.get_balance())
```

ques.52 Modify the decorator to accept arguments and print the function name along with the message?

```python
from functools import wraps

def log_execution(message):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            print(f"{message} - Before executing {func.__name__}")
            result = func(*args, **kwargs)
            print(f"{message} - After executing {func.__name__}")
            return result
        return wrapper
    return decorator

@log_execution("LOG")
def say_hello():
    print("Hello, world!")
say_hello()
```

ques.62 How does the __len__() method work in Python? Provide an example?

The __len__() method in Python is a special method used to define the behavior of the built-in len() function for a custom class.

.When you call len(obj), Python internally calls obj.__len__() and returns an integer.

.This allows your objects to behave like sequences or collections (lists, strings, etc.) in terms of length.

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)

team = Team(["Alice", "Bob", "Charlie"])
print(len(team))
```

ques.63 Explain the usage of the __add__() method in Python classes. Provide an example?

The __add__() method in Python is a special (magic) method that defines the behavior of the + operator for custom objects.

.When you do obj1 + obj2, Python internally calls obj1.__add__(obj2).

.You can override it in your class to define how addition should work between instances of your class.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Point):
            return NotImplemented

        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Point({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2
print(p3)
```

ques.64 What is the purpose of the __getitem__() method in Python? Provide an example?

The __getitem__() method in Python is a special (magic) method that allows a class to support indexing and key access using square brackets ([]).

.When you do obj[key], Python internally calls obj.__getitem__(key).

.This makes your objects behave like lists, dictionaries, or other collections.

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __getitem__(self, index):
        return self.members[index]

team = Team(["Alice", "Bob", "Charlie"])

print(team[0])
print(team[2])
```

.Raises StopIteration when there are no more items.

```python
class CountUpTo:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        # The object itself is the iterator
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

counter = CountUpTo(5)

for num in counter:
    print(num)
```

ques.66 What is the purpose of a getter method in Python? Provide an example demonstrating the use of a getter
method using property decorators?

A getter method in Python is used to access (read) the value of an attribute in a controlled way.

.Enforce encapsulation

.Add validation or transformation when reading a value

.Keep a stable public interface even if internal implementation changes

.Make a computed value look like a normal attribute

```python
class Person:
    def __init__(self, name, birth_year):
        self._name = name
        self._birth_year = birth_year

    @property
    def name(self):
        return self._name

    @property
    def age(self):
        return 2026 - self._birth_year

p = Person("Alex", 1995)

print(p.name)
print(p.age)
```

ques.67 Explain the role of setter methods in Python. Demonstrate how to use a setter method to modify a class
attribute using property decorators?

A setter method in Python is used to control how an attribute’s value is modified.

.Enforce validation rules

.Maintain data integrity

.Trigger side effects (logging, recalculation, notifications)

.Preserve encapsulation while keeping clean attribute-style access

```python
class BankAccount:
    def __init__(self, owner, balance):
        self._owner = owner
        self._balance = balance

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, amount):
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = amount

account = BankAccount("Jordan", 1000)

account.balance = 1500
print(account.balance)

```

ques.68 What is the purpose of the @property decorator in Python? Provide an example illustrating its usage?

The @property decorator in Python is used to turn a method into a managed attribute.

.Clean, attribute-style access

.Encapsulation and data protection

.Logic (validation, computation, formatting) without changing how the attribute is used

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property
    def area(self):
        return self._width * self._height

rect = Rectangle(4, 5)

print(rect.area)
```

ques.69 Explain the use of the @deleter decorator in Python property decorators. Provide a code example
demonstrating its application?

The @deleter decorator in Python is used with @property to define what happens when an attribute is deleted using the del keyword.

.Perform cleanup actions (close files, release resources)

.Reset related internal state

.Prevent accidental deletion by raising an error

.Log or audit deletion events

```python
class User:
    def __init__(self, username):
        self._username = username

    @property
    def username(self):
        return self._username

    @username.setter
    def username(self, value):
        if not value:
            raise ValueError("Username cannot be empty")
        self._username = value

    @username.deleter
    def username(self):
        print("Deleting username...")
        del self._username

user = User("alice")

print(user.username)

del user.username
```

ques.70 How does encapsulation relate to property decorators in Python? Provide an example showcasing
encapsulation using property decorators?

Encapsulation is the object-oriented principle of hiding internal data and exposing controlled access to it.
In Python, property decorators (@property, @setter, @deleter) are a key way to achieve encapsulation while keeping code clean and intuitive.

    Property decorators help you:

.Hide internal attributes (often prefixed with _)

.Control how data is read, modified, or deleted

.Prevent invalid or inconsistent state

.Change internal implementation without breaking external code

```python
class Employee:
    def __init__(self, name, salary):
        self._name = name
        self._salary = salary

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative")
        self._salary = value

emp = Employee("Sam", 50000)

print(emp.salary)

emp.salary = 60000

```
