ques.16 What is REST, and what are its key principles?

REST (Representational State Transfer) is an architectural style for designing networked applications, especially Web APIs. It’s not a protocol or standard itself—more like a set of guidelines and principles that make APIs scalable, reliable, and easy to use.

1. Statelessness

.Each request from a client contains all the information needed for the server to process it.

.The server does not store session information between requests.

.Benefit: Makes the API scalable and easier to cache.

2. Client-Server Architecture

.Client (frontend, mobile app) and server (API/backend) are separate.

.The client handles user interface and experience; the server handles data and business logic.

.Benefit: Each side can evolve independently.

3. Layered System

.REST APIs can have multiple layers (load balancers, caches, proxies) between client and server.

.Each layer is independent and unaware of the others’ details.

.Benefit: Improves scalability, reliability, and security.

4. Cacheability

.Responses must indicate whether they are cacheable (Cache-Control, ETag).

.Clients or intermediaries can store responses to reduce repeated work.

.Benefit: Reduces latency and server load.

ques.18  What are the main HTTP methods used in RESTful architecture, and what are their purposes?


In RESTful architecture, HTTP methods are the verbs that indicate what action the client wants to perform on a resource.

1. GET - Retrieve a resource

.Purpose: Read or fetch data without modifying it.

.Safe & idempotent: Does not change server state; repeated calls have the same effect.

   .Use cases:

.Fetch a single resource: GET /users/42

.Fetch a collection: GET /users

.Responses: Usually 200 OK with JSON (or XML) body.

2. POST - Create a new resource

.Purpose: Send data to the server to create something new.

.Not idempotent: Multiple identical POST requests can create multiple resources.

   .Use cases:

.Create a new user: POST /users with JSON body { "name": "Alice" }

.Trigger a server-side action that produces a new resource

.Responses: 201 Created (often with Location header pointing to the new resource)

3. PUT - Update/replace a resource

.Purpose: Replace the entire resource at a given URI.

.Idempotent: Multiple identical PUT requests have the same effect as one.

   .Use cases:

.Update user info: PUT /users/42 with full JSON object

.Responses: 200 OK or 204 No Content if no body is returned

ques.20 What is the significance of URIs (Uniform Resource Identifiers) in RESTful API design?

URIs (Uniform Resource Identifiers) are fundamental in RESTful API design because they define the “things” your API exposes and determine how clients interact with them.

1. URIs identify resources

.In REST, everything that clients can interact with is a resource: users, orders, products, etc.

.A URI uniquely identifies each resource or collection of resources.

   Example:

  /users          → collection of users
  /users/42       → specific user with ID 42
  /orders/2021    → orders from year 2021


Without clear URIs, clients cannot reliably locate or manipulate resources.

2. URIs make APIs predictable

.Consistent, logical URIs allow developers to guess and navigate your API easily.

.They help ensure a uniform interface, one of the core REST principles.

   Good practice:

.Use nouns, not verbs: /orders instead of /getOrders

.Use plural nouns for collections: /products instead of /product

.Represent hierarchy through path segments: /users/42/orders/7

3. URIs enable statelessness

.Because a URI fully identifies a resource, the server does not need to remember any previous state about the client.

.Every request specifies the resource explicitly via its URI, often combined with query parameters or headers.

ques.21  Explain the role of hypermedia in RESTful APIs. How does it relate to HATEOAS?


Hypermedia plays a central role in making a RESTful API truly RESTful. It is the mechanism that allows clients to dynamically discover available actions and navigate an application purely through links provided by the server.

Hypermedia is an extension of hypertext.
It means that responses include:

.Data (resource representation)

.Links (URIs to related resources)

.Possible actions (what the client can do next)

Instead of just returning raw JSON data, a hypermedia-driven API includes links inside the response.



```python
{
  "id": 123,
  "status": "pending",
  "total": 49.99,
  "links": [
    { "rel": "self", "href": "/orders/123" },
    { "rel": "cancel", "href": "/orders/123/cancel" },
    { "rel": "pay", "href": "/orders/123/payment" }
  ]
}
```

HATEOAS stands for:

Hypermedia As The Engine Of Application State

It means:

.The client interacts with the application entirely through hypermedia links provided by the server

.The server drives state transitions

.The client does not construct URLs manually

The client starts from an entry point (e.g., /api) and navigates the API using links returned in responses.

ques.23 Discuss the concept of resource representations in RESTful APIs?

In RESTful APIs, resource representations are the concrete, transferable forms of a resource’s state sent between client and server.

1. Resource vs. Representation

A core idea introduced by Roy Fielding in his dissertation on REST is the separation between:

.Resource → A conceptual target identified by a URI

.Representation → A snapshot of that resource’s state in a specific format

 For example:

 GET /users/123
 Accept: application/json

2. Representations and HTTP Methods

Representations play different roles depending on the HTTP method:

.GET → Server returns a representation of the resource.

.POST → Client sends a representation to create a new resource.

.PUT → Client sends a complete representation to replace a resource.

.PATCH → Client sends a partial representation to update part of a resource.

ques.32 Describe the structure of a SOAP message?

A SOAP message is an XML document with a strict, standardized structure defined by the SOAP specification (standardized by the World Wide Web Consortium).

High-Level Structure of a SOAP Message

 A SOAP message has four main parts:

.Envelope (required)

.Header (optional)

.Body (required)

.Fault (optional, inside Body)

1. SOAP Envelope (Required)

 The <Envelope> element is the root element of the SOAP message.

 It:

.Identifies the XML as a SOAP message

.Defines namespaces

.Contains Header and Body elements

2. SOAP Header (Optional)

The <Header> contains metadata about the message.

 It is commonly used for:

.Authentication credentials

.Security tokens

.Transaction IDs

.Routing information

.Digital signatures

3. SOAP Body (Required)

 The <Body> contains the actual message payload.

 This includes:

.The request (for client → server)

.The response (for server → client)

4. SOAP Fault (Optional, Inside Body)

 If an error occurs, SOAP returns a <Fault> element inside the Body.

 Structure of a SOAP Fault:

.Code - Error type

.Reason - Human-readable message

.Detail - Additional error details

SOAP (Simple Object Access Protocol) is a protocol for exchanging structured information between clients and servers in web services. It relies on XML and typically runs over standard application protocols like HTTP or SMTP.

.It generates a SOAP response.
